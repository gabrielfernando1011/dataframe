```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('vendas.xlsx')
display(df)



```

```python
# 2. Padronizar a coluna 'categoria' para minúsculas
df['categoria'] = df['categoria'].str.lower()
print(df)
```

```python
# 3. Remover registros onde a quantidade está ausente
df = df.dropna(subset=['quantidade'])
print(df)
```

```python
# 4. Remover espaços extras e padronizar para estilo título
df['loja'] = df['loja'].str.strip().str.title()
df['categoria'] = df['categoria'].str.strip().str.title()
df['produto'] = df['produto'].str.strip().str.title()
print(df)
```

```python
# 5. Garantir que a coluna 'data' esteja no formato datetime
df['data'] = pd.to_datetime(df['data'], errors='coerce')
print(df)
```

```python
# 6. Calcular valor unitário real (valor_venda / quantidade)
df['valor_unitario'] = df['valor_venda'] / df['quantidade']
print(df)
```

```python
# 7. Calcular valor médio unitário por produto (sem considerar nulos)
media_unitaria_produto = df.groupby('produto')['valor_unitario'].mean()
print(df)
```

```python

# 8. Preencher valores ausentes em valor_unitario com a média do produto
df['valor_unitario'] = df.apply(
    lambda row: media_unitaria_produto[row['produto']] if pd.isna(row['valor_unitario']) else row['valor_unitario'],
    axis=1
)
print(df)
```

```python
# 9. Recalcular valor_venda com base em valor_unitario * quantidade
df['valor_venda'] = df['valor_unitario'] * df['quantidade']
print (df)
```

```python
# 10. Salvar o novo DataFrame limpo em um novo arquivo Excel
df.to_excel('vendas_limpo.xlsx', index=False)

print("Arquivo 'vendas_limpo.xlsx' salvo com sucesso!")
```

```python
# Carregar a planilha
file_path = "vendas_limpo.xlsx"  # Altere se o arquivo estiver em outro local
df = pd.read_excel(file_path)
print(df)
```

```python
# Converter a coluna de data para formato datetime
df["data"] = pd.to_datetime(df["data"])
df["ano_mes"] = df["data"].dt.to_period("M")
print(df)
```

```python
# 1. Total de vendas por loja
vendas_por_loja = df.groupby("loja")["valor_venda"].sum().sort_values()
print("\nTotal de vendas por loja:\n", vendas_por_loja)
vendas_por_loja.plot(kind='bar', figsize=(10,6),colormap='Set2')

# Configurações do gráfico
plt.title("Total de Vendas por lojas")
plt.xlabel("Lojas")
plt.ylabel("Total das Vendas")
plt.legend(title="Categoria")
plt.xticks(rotation=90)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar o gráfico
plt.show()
```

```python
# 2 e 3. Evolução mensal de vendas
evolucao_mensal = df.groupby("ano_mes")["valor_venda"].sum().sort_values()
print("\nEvolução mensal de vendas:\n", evolucao_mensal)
evolucao_mensal.plot(kind='bar', figsize=(10,6),colormap='Set2')
# Configurações do gráfico
plt.title("Evolucao Mensal por Mes")
plt.xlabel("Ano/Mes")
plt.ylabel("Valores de Vendas")
plt.legend(title="Categoria")
plt.xticks(rotation=90)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar o gráfico
plt.show()
```

```python
# 4. Produtos mais vendidos (por volume)
produtos_mais_vendidos = df.groupby("produto")["quantidade"].sum().sort_values()
print("\nProdutos mais vendidos:\n", produtos_mais_vendidos.head())
produtos_mais_vendidos.plot(kind='bar', figsize=(10,6),colormap='Set2')

# Configurações do gráfico
plt.title("Produtos mais Vendidos (por volume)")
plt.xlabel("Produtos")
plt.ylabel("Quantidade")
plt.legend(title="Categoria")
plt.xticks(rotation=90)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar o gráfico
plt.show()
```

```python
# 5. Comparação entre categorias
vendas_por_categoria = df.groupby("categoria")["valor_venda"].sum().sort_values()
print("\nVendas por categoria:\n", vendas_por_categoria)
vendas_por_categoria.plot(kind='bar', figsize=(10,6),colormap='Set2')

# Configurações do gráfico
plt.title("Comparacao por Categoria")
plt.xlabel("Categorias")
plt.ylabel("Soma dos Valores")
plt.legend(title="Categoria")
plt.xticks(rotation=90)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar o gráfico
plt.show()
```

```python
# 6. Comparação entre lojas por categoria
vendas_loja_categoria = df.groupby(["loja", "categoria"])["valor_venda"].sum().unstack()
print(df)
```

```python
# 7. Gerar gráfico de barras empilhadas

vendas_loja_categoria.plot(kind="bar", figsize=(10,6), stacked=True, colormap="Set2")

# Configurações do gráfico
plt.title("Comparação de Vendas por Categoria e Loja")
plt.xlabel("Lojas")
plt.ylabel("Valor Total de Vendas (R$)")
plt.legend(title="Categoria")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar o gráfico
plt.show()
```
